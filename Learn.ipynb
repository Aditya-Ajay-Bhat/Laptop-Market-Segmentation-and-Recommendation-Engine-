```python


import pandas as pd
import numpy as np

# Create a boolean mask to find rows where the 'Core' column contains RAM-related text.
# The `|` symbol acts as an "OR", so we're searching for "RAM" OR "Memory".
# `na=False` ensures that actual missing values (NaNs) are treated as False.
# `case=False` makes the search case-insensitive.
crap_core_mask = df['Core'].str.contains('RAM|Memory', case=False, na=False)

# Filter your df1 DataFrame using this mask
incorrect_core_rows = df[crap_core_mask]

# Display the resulting table of incorrect rows
print(f"Found {len(incorrect_core_rows)} rows where the 'Core' column contains RAM or Memory specs:")
incorrect_core_rows["Model"]
```

```python
import pandas as pd

# This is the list of incorrect values you want to find in the 'Display' column.
crap_display_values = [
    'Windows 11 OS',
    '4\u2009GB NVIDIA GeForce RTX 3050 ',
    '1 Year Warranty',
    'Display: 15.6 inch (39.6 cm)',
    '4\u2009GB NVIDIA GeForce RTX 2050 ',
    'Windows OS',
    '4\u2009GB \u200eNVIDIA Quadro T600',
    'Windows 11  OS'
]

# Create a boolean filter to find all rows where the 'Display' value is in our list.
incorrect_display_mask = df['Display'].isin(crap_display_values)

# Apply the filter to the df1 DataFrame
incorrect_display_rows = df[incorrect_display_mask]

# Display the resulting table of incorrect rows.
print(f"Found {len(incorrect_display_rows)} rows with incorrect 'Display' values:")
incorrect_display_rows["Model"]
```
